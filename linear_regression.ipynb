```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('medical_expenses.csv')

X = dataset.iloc[:,0]
Y = dataset.iloc[:,1]
```

```python
#y = mx + c
xy = X*Y
xy_mean = xy.mean()
x_mean = X.mean()
y_mean = Y.mean()
x_square = X*X
x_square_mean = x_square.mean()
x_mean_square = x_mean * x_mean
```

```python
m = (xy_mean - (x_mean * y_mean)) / (x_square_mean - x_mean_square)

c = y_mean - (m * x_mean)

print( f'm = {m}    c = {c}')
```

```python
y_hat = (m * X ) + c
y_hat
```

```python
plt.scatter(X, Y, color='blue', label='Data Points')  # Scatter plot of data
plt.plot(X, y_hat, color='green', label='Best Fit Line')  # Line of best fit
plt.xlabel('X (Independent Variable)')
plt.ylabel('Y (Dependent Variable)')
plt.title('Linear Regression Line of Best Fit')
plt.legend()
plt.show()
```

```python
#loss function
#find best m and minimize the loss function

M = 0
C = 0
alpha = 0.00001
iteration = 1000 #no of iteration
n = len(X)



for iteration in range(iteration):
     Y_hat = M * X + C

     dm = -2/n * np.sum(X * (Y - Y_hat))
     dc = -2/n * np.sum(Y - Y_hat)

     M = M - (alpha * dm)
     C = C - (alpha * dc)

     loss = 1/n * np.sum((Y - Y_hat)**2)

    

     if iteration % 100 == 0:
        print(f"Iteration {iteration}: Loss = {loss}, m = {m}, c = {c}")
    
    

print(f'm = {M}, c = {C}')




```

```python
Updated_y_hat = (M * X ) + C
Updated_y_hat
```

```python
plt.scatter(X, Y, color='blue', label='Data Points')  # Scatter plot of data
plt.plot(X, y_hat, color='green', label=' regression line') 
plt.plot(X, Updated_y_hat, color='red', label='Best Fit Line')  # Line of best fit
plt.xlabel('X (Independent Variable)')
plt.ylabel('Y (Dependent Variable)')
plt.title('Linear Regression Line of Best Fit')
plt.legend()
plt.show()
```
